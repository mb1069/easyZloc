```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# image inputs
img_train = train_dataset.data['train'][0][0].squeeze()
img_test = exp_dataset.data['all'][0][0].squeeze()

print(img_train.shape)
print(img_test.shape)

def img_stats(imgs, name):
    return {
        'peak': np.max(imgs, axis=(1, 2)),
        'std': np.std(imgs, axis=(1, 2)),
        'mean': np.mean(imgs, axis=(1, 2)),
        'min': np.min(imgs, axis=(1, 2)),
        'median': np.median(imgs, axis=(1, 2)),
        'type': [name] * len(imgs)
    }

train_img_stats = img_stats(img_train, 'train')
test_img_stats = img_stats(img_test, 'test')

dfs = [pd.DataFrame.from_dict(img_stats) for img_stats in [train_img_stats, test_img_stats]]
df = pd.concat(dfs)

df.boxplot(by='type', column='mean')
df.boxplot(by='type', column='peak')
df.boxplot(by='type', column='median')
df.boxplot(by='type', column='min')
df.boxplot(by='type', column='std')
plt.show()
distances_train = train_dataset.data['train'][0][1].squeeze()
distances_test = exp_dataset.data['all'][0][1].squeeze()


train_df = train_dataset.csv_data
exp_df = exp_dataset.csv_data

plt.scatter(train_df['x [nm]'], train_df['y [nm]'], label='train')
plt.scatter(exp_df['x [nm]'], exp_df['y [nm]'], label='exp')
plt.legend()
plt.show()
```
